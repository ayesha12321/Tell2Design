```python
import tqdm
from torch.utils.data import DataLoader
from transformers.generation import StoppingCriteria, StoppingCriteriaList


SCAD_OUTPUT_DIR = "./"



if not os.path.exists(SCAD_OUTPUT_DIR):
    os.makedirs(SCAD_OUTPUT_DIR)

def prompt_input_sentence_with_boundary(data_point):
  eval_prompt = f"""You are a powerful text-to-SCAD model. Your job is to generate SCAD code based on natural language input.

### Input:
{data_point["sentence"] + data_point['boundary_tokens']}

### Response:
"""
  return eval_prompt.format(data_point=data_point)



def collate_fn(batch):
    image_ids = [row["image_id"] for row in batch]
    # sentences = [prompt_input_sentence(row["sentence"]) for row in batch]
    sentences = [prompt_input_sentence_with_boundary(row) for row in batch]
    scads = [row["scad"] for row in batch]
    return image_ids, sentences, scads

batch_size = 1

test_dataloader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False, collate_fn=collate_fn)

tokenizer.pad_token = tokenizer.eos_token
tokenizer.padding_side = "left"

def write_scad_file(scad_code, filename):
    with open(filename, 'w', encoding='utf-8') as file:
        file.write(scad_code)


def check_file_exists(file_path):
    if os.path.isfile(file_path):
        print(f"Image file '{file_path}' exists.")
        return True
    else:
        # print(f"Image file '{file_path}' does not exist.")
        return False




stop_string = "<|theend|>"


results = []
for image_ids, sentences, scads in tqdm.tqdm(test_dataloader):

    infer = False

    if image_ids[0] != '3908.png':
      continue

    for image_id in image_ids:
      scad_name = image_id.replace("png","scad")
      final_file = f"{SCAD_OUTPUT_DIR}/{scad_name}"
      if check_file_exists(final_file):
        continue
      else:
        infer = True
        break

    if not infer:
      continue


    tokenized_inputs = tokenizer(sentences, truncation=True,
            max_length=1024,
            padding=True,
            return_tensors="pt",
    ).to("cuda")


    with torch.no_grad():
        generated_tokens = trained_model.generate(
            **tokenized_inputs,
            max_new_tokens=800,
            pad_token_id=tokenizer.eos_token_id,
        )

    decoded_outputs = tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)

    post_processed_decoded_outputs = []
    for output in decoded_outputs:
        output_post = output.split(stop_string)[0].strip()
        post_processed_decoded_outputs.append(output_post)

    for image_id, sentence, scad, full_output in zip(image_ids, sentences, scads, post_processed_decoded_outputs):
        scad_name = image_id.replace("png","scad")
        output = full_output[len(sentence):].strip()


        write_scad_file(output, f"{SCAD_OUTPUT_DIR}/{scad_name}")
        print("scad_name:", scad_name)


```

```python
import re
from collections import Counter, defaultdict

def extract_from_polygons(code_string, type='BBOX'):




    color_regex = r'color\("#([A-Fa-f0-9]{6})"\)'
    colors = re.findall(color_regex, code_string)

    polygon_regex = r'polygon\(\[([\[\d+,\s\]]+)\]\);'
    polygons = re.findall(polygon_regex, code_string)


    polygons = [p.strip() for p in polygons]

    bounding_boxes = defaultdict(tuple)
    color_count = defaultdict(int)
    processed_colors = []
    for color in colors:
        color_count[color] += 1
        if color_count[color] > 1:
            processed_colors.append(f"{color}_{color_count[color]-1}")
        else:
            processed_colors.append(color)


    for color, polygon in zip(processed_colors, polygons):
        coords = eval(polygon)

        if type == "BBOX":
            x_coords = [x[1] for x in coords]
            y_coords = [x[0] for x in coords]
            xmin, xmax = min(x_coords), max(x_coords)
            ymin, ymax = min(y_coords), max(y_coords)

            bounding_boxes[color] = (ymin, xmin, ymax, xmax)
        else:
            bounding_boxes[color] = coords

    return bounding_boxes

import cv2
import numpy as np
import matplotlib.pyplot as plt

def draw_polygons_with_colors(image_size, polygons):
    """
    Draws polygons with respective colors on a white image.

    Args:
        image_size (tuple): Size of the image (height, width).
        polygons (dict): Dictionary where keys are hex color names (may include _1, _2) and
                         values are lists of polygonal points.

    Returns:
        numpy.ndarray: Image with polygons drawn in respective colors.
    """
    image = np.ones((image_size[1], image_size[0], 3), dtype=np.uint8) * 255

    for color, points in polygons.items():
        base_color = color.split('_')[0]

        bgr_color = tuple(int(base_color[i:i+2], 16) for i in (4, 2, 0))

        points = np.array([tuple(point) for point in points], dtype=np.int32)

        cv2.fillPoly(image, [points], bgr_color)

    return image

image_size = (256, 256)

```
